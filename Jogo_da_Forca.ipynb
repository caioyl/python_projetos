```python
import random
```

```python
lista_palavras = ['python','javascript','django']
palavra = random.choice(lista_palavras)
letras_usuario = []
chances = 4

while True:
  #Itera as letras da palavra escolhida e as imprime caso estejam contidas na lista das letras digitadas pelo usuário. Caso contrário, imprime underline
  for letra in palavra:
    if letra.lower() in letras_usuario:
      print(letra,end=' ')
    else:
      print('_',end=' ')
  #Imprime a quantidade de chances restantes
  print(f'Chances restantes: {chances}')
  #Recebe o vlaor da tentativa
  tentativa = input('Escolha uma letra para adivinhar: ')
  #Adiciona a tentativa à lista de letras tecladas pelo usuário
  letras_usuario.append(tentativa.lower())
  #Controle de chances: se errar, perde uma chance
  if tentativa.lower() not in palavra.lower():
    chances -= 1
  #Define ganhou como True a cada laço
  ganhou = True
  #Define ganhou como falso caso alguma letra da palavra não esteja contida nas letras tecladas pelo usuário
  for letra in palavra:
    if letra.lower() not in letras_usuario:
      ganhou = False

  #Encerramento
  if chances == 0 or ganhou:
    break


if ganhou:
  print(f'Parabéns, você ganhou! A palavra era: {palavra}')
else:
  print(f'Você perdeu! A palavra era: {palavra}')
```
